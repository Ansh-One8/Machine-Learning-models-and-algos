### LINEAR REGRESSION

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

A house with 1000 square feet(sqft) sold for $300,000 and a house with 2000 square feet sold for $500,000. These two points will constitute our data or training set. In this lab, the units of size are 1000 sqft and the units of price are 1000s of dollars

```python
#SO, we have two house's data 
## let this x_train data be sqr feet data in 1000sq
## let this y_train data be of prize of the houses in 1000s of dollar
x_train = np.array([1.0 , 2.0])
y_train = np.array([300.0 , 500.0])
print(f" x_train = {x_train}")
print(f" y_train = {y_train}")
```

```python
#to get m--> m is the number of training examples
m = len(x_train)
print(f" m is the number of training examples  = {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train,y_train, c = 'r')
plt.xlabel('sqaure feet in 1000s')
plt.ylabel('price')
plt.title("house predict point hehehehehe")
```

model rep start

```python
w  = 200
b = 100
def compute_model_output(x,w,b) :
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
       f_wb[i] = w * x[i] + b
    return f_wb
    
```

```python
tmp_fwb = compute_model_output(x_train,w,b )

plt.plot(x_train , tmp_fwb , c = 'b')

plt.scatter(x_train , y_train , c ='r')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
