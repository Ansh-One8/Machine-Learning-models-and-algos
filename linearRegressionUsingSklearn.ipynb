```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
#this is not a running code but trust me ..this how the iteration happen using sklearn 
```

```python
# suppose theres a house dataset , to load dataset 
X_train, y_train = load_house_data()
```

LinearRegression — just call .fit() directly on raw data, done in one step.
SGDRegressor — always normalize first with StandardScaler, then fit, and remember to use scaler.transform() (not fit_transform) on test data too. 
That's the extra step so dont  forget!hehehehe

```python
# now as you know x_train would have many feature, and you need to normalise it as age and bedroom or size would be in different way so better to normalise
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)
print(f"Peak to Peak range by column in Raw        X:{np.ptp(X_train,axis=0)}")   
print(f"Peak to Peak range by column in Normalized X:{np.ptp(X_norm,axis=0)}")
#important point:- here fit_transforn does two thing 1) fit the dataset it learns for example;-it learns the value of mean of size , std of bedroom ,etc
# 2) but here Transform:-scale the data of mean of size or std bedroom 
```

Training data → fit_transform ✅
# X_train_norm = scaler.fit_transform(X_train)

Test data → only transform ✅ NEVER fit_transform!
# X_test_norm = scaler.transform(X_test)

fit learns the mean and std from training data and 
Test data should use the same mean and std as training data.....
If you fit on test data → you're cheating by letting model peek at test data!

Simple analogy:

fit = a tailor measuring your body
transform = cutting the cloth based on those measurements
fit_transform = measuring AND cutting in one step!



```python
#CREATE AND FIT THE REGRESSION MODEL
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(X_norm, y_train)
print(sgdr)
print(f"number of iterations completed: {sgdr.n_iter_}, number of weight updates: {sgdr.t_}")
#HERE ,
```

```python
b_norm = sgdr.intercept_
w_norm = sgdr.coef_
print(f"model parameters:                   w: {w_norm}, b:{b_norm}")
print( "model parameters from previous lab: w: [110.56 -21.27 -32.71 -37.97], b: 363.16")
```

```python
# make a prediction using sgdr.predict()
y_pred_sgd = sgdr.predict(X_norm)
# make a prediction using w,b. 
y_pred = np.dot(X_norm, w_norm) + b_norm  
print(f"prediction using np.dot() and sgdr.predict match: {(y_pred == y_pred_sgd).all()}")

print(f"Prediction on training set:\n{y_pred[:4]}" )
print(f"Target values \n{y_train[:4]}")
# next step is to plot as it is  just a notes more than a running code , so , i wont do it if anyone thinking looking at my github hii btw!!!
# yo byez xoxo 
```
